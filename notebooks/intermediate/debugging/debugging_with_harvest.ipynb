```python
import jax
import jax.numpy as jnp
import genjax
from genjax import tag, grab, stab

console = genjax.pretty()
```

```python
def f(x):
    x = tag(x + 1, name="x")
    result = tag(x**2, name="result")
    return result


v, state = genjax.grab(f)(5.0)
state
```

```python
v, state = stab(f)({"x": 7.0}, (5.0,))
state
```

```python
@genjax.gen
def model():
    z = genjax.Normal(0.0, 1.0) @ "z"
    tag(z, name="z")
    x = genjax.Normal(z, 1.0) @ "x"
    return x
```

```python
key = jax.random.PRNGKey(314159)
constraints = genjax.choice_map({"x": 2.0})
inf = genjax.ImportanceSampling(100, model)
(key, v), state = grab(inf.apply)(key, constraints, ())
state
```
